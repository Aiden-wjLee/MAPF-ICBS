```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# cbs_st = pd.read_csv('nodes-gen-cleaned.csv', names=['instance', 'nodes'])


# all nodes generated

cbs_st_gen = pd.read_csv('data/CBS/nodes-gen-cleaned.csv', names=['instance', 'nodes'])
cbs_di_gen = pd.read_csv('data/CBS--disjoint/nodes-gen-cleaned.csv', names=['instance', 'nodes'])

icbscb_st_gen = pd.read_csv('data/ICBS_CB/nodes-gen-cleaned.csv', names=['instance', 'nodes'])
icbscb_di_gen = pd.read_csv('data/ICBS_CB--disjoint/nodes-gen-cleaned.csv', names=['instance', 'nodes'])

icbs_st_gen = pd.read_csv('data/ICBS/nodes-gen-cleaned.csv', names=['instance', 'nodes'])
icbs_di_gen = pd.read_csv('data/ICBS--disjoint/nodes-gen-cleaned.csv', names=['instance', 'nodes'])

# all nodes expanded
cbs_st_exp = pd.read_csv('data/CBS/nodes-exp-cleaned.csv', names=['instance', 'nodes']) 
cbs_di_exp = pd.read_csv('data/CBS--disjoint/nodes-exp-cleaned.csv', names=['instance', 'nodes']) 

icbscb_st_exp = pd.read_csv('data/ICBS_CB/nodes-exp-cleaned.csv', names=['instance', 'nodes'])
icbscb_di_exp = pd.read_csv('data/ICBS_CB--disjoint/nodes-exp-cleaned.csv', names=['instance', 'nodes'])

icbs_st_exp = pd.read_csv('data/ICBS/nodes-exp-cleaned.csv', names=['instance', 'nodes'])
icbs_di_exp = pd.read_csv('data/ICBS--disjoint/nodes-exp-cleaned.csv', names=['instance', 'nodes'])

headers = ['CBS + Standard','CBS + Disjoint','ICBS_CB + Standard','ICBS_CB + Disjoint','ICBS + Standard','ICBS + Disjoint']

#all nodes generated df
gen_data = [cbs_st_gen['nodes'],cbs_di_gen['nodes'],icbscb_st_gen['nodes'],icbscb_di_gen['nodes'],icbs_st_gen['nodes'],icbs_di_gen['nodes']]

exp_data = [cbs_st_exp['nodes'],cbs_di_exp['nodes'],icbscb_st_exp['nodes'],icbscb_di_exp['nodes'],icbs_st_exp['nodes'],icbs_di_exp['nodes']]

gen_batch_df = pd.concat(gen_data, axis=1, keys=headers)
exp_batch_df = pd.concat(exp_data, axis=1, keys=headers)
# exp_df
```

```python
# gen_batch_df

expsorted_batch_df = exp_batch_df.sort_values(by=['CBS + Standard'])

expsorted_batch_df = expsorted_batch_df.reset_index(drop=True)

expsorted_batch_df
```

```python
# For the batch

sns.set_style("whitegrid")
fig, ax = plt.subplots()

sns.set(rc={'figure.figsize':(12,9)})

# Plot nodes generated
# sns.set_color_codes('pastel')
sns.barplot(data=gen_batch_df,
            label='Nodes generated', color='lightblue')

# Plot nodes expanded
# sns.set_color_codes('muted')
sns.barplot(data=exp_batch_df,
            label='Nodes expanded', color='black')
ax.set_ylabel('')


ax.set_yscale("log")
ax.legend(ncol=2, loc="upper right", frameon=True)
ax.set(xlabel='Average nodes generated and nodes expanded: batch test')
sns.despine(left=True, bottom=True)
# plt.savefig('ten_most_reviewed.png', bbox_inches='tight')
```
